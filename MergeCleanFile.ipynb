```python
import pandas as pd
import seaborn as sns
import matplotlib as plt
```

```python

df = pd.read_csv(r"https://raw.githubusercontent.com/mredshaw/AI-Fundamentals-Project/main/CleanData.csv")
```

```python
df
```

```python
# Define the list of columns to update
columns_to_update = ['1039','1040','1041','1042','1043','1044','1044','1045','1046','1047','1048','1049','1050','1051','1052','1053']

# Change the value in the first row of specified columns to a hard-coded float value
for column in columns_to_update:
    df.at[0, column] = 6.9
```

```python
def convert_to_float(value):
    if value == 'Characteristic':
        return value
    else:
        return (float(value))

# Apply the function to all elements in the DataFrame except the first row and first column
df.iloc[1:, 1:] = df.iloc[1:, 1:].applymap(convert_to_float)
```

```python
columns_to_merge = ['1039','1040','1041','1042','1043','1044','1044','1045','1046']
# Calculate the sum of the specified columns
merged_column_sum = df[columns_to_merge].sum(axis=1)

# Create a new column with the merged values
merged_column_name = 'Under $50,000'
df[merged_column_name] = merged_column_sum

# Remove the individual columns that were merged
df = df.drop(columns_to_merge, axis=1)
```

```python
columns_to_merge = ['1047','1048','1049','1050','1051']
# Calculate the sum of the specified columns
merged_column_sum = df[columns_to_merge].sum(axis=1)

# Create a new column with the merged values
merged_column_name = '$50,000 - $100,000'
df[merged_column_name] = merged_column_sum

# Remove the individual columns that were merged
df = df.drop(columns_to_merge, axis=1)
```

```python
column_mapping = {'1052': '$100,000 and over','1053': '$200,000 and over'}
df = df.rename(columns=column_mapping)
```

```python
columns_to_shift = ['$100,000 and over', '$200,000 and over']

# Create a new DataFrame with the specified columns moved to the end
shifted_df = df[[col for col in df.columns if col not in columns_to_shift] + columns_to_shift]

# Assign the new DataFrame back to the original DataFrame
df = shifted_df
```

```python
df.to_csv('TestCleanData.csv')
```
