```python
df2= df_transposed
```

```python
def remove_comma_and_convert_to_float(value):
    if isinstance(value, str):
        return float(value.replace(",", ""))
    else:
        return value

# Apply the function to all elements in the DataFrame except the first row
df2.iloc[1:] = df2.iloc[1:].applymap(remove_comma_and_convert_to_float)
```

```python
df2.to_csv('CleanData.csv')
```
